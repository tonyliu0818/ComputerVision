```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
from scipy import linalg
%matplotlib inline

def show_gradient_descent(A, b):
#  enter your code here, and remove the following line (30%)
   #pass
   x=[[-2],[2]]
   x0,x1,f_=[],[],[]
   x0.append(x[0][0])
   x1.append(x[1][0])
   alpha,iter=0.02,1000
   f=np.linalg.norm(A@x-b)**2
   current_slope=[]
   current_slope.append(f)
   for k in range(iter):
     print(f"Epochs:{k},(x1,x2):({x[0][0]:.4f},{x[1][0]:4f}),f(x):{f:.8f}")
     x_prev=x
     deltaf=2*(A.T@A@x-A.T@b)
     x=x-alpha*deltaf
     if np.linalg.norm(x-x_prev)<=1e-8:
       break
     f=np.linalg.norm(A@x-b)**2
     x0.append(x[0][0])
     x1.append(x[1][0])
     current_slope.append(f)
   return x0,x1,current_slope   
```

```python
def draw_gradient_descent(A, b,x0,x1,slope):
  xx, yy = np.meshgrid(np.arange(-2,2.1,0.1), np.arange(-2,2.1,0.1))
  F1 = A[0][0]*xx + A[0][1]*yy - b[0];
  F2 = A[1][0]*xx + A[1][1]*yy - b[1];
  zz = F1*F1 + F2*F2;
  plt3d = plt.figure().add_subplot(projection='3d')
  plt3d.plot_surface(xx, yy, zz, alpha=0.7, cmap=cm.coolwarm)
    
#  enter your code here (14%)
  ax=plt.gca()
  ax.scatter(x0,x1,slope,c='r')
  x=np.arange(-2,2.1,0.1)
  #y-y1=m(x-x1)
  y=slope[-1]*(x-x0[-1])+x1[-1]
  ax.plot(x,y,c='b')
  #limit x and y
  ax.set_xlim((-2.1,2.1))
  ax.set_ylim((-2.1,2.1))
  ax.view_init(elev=10,azim=20)
  #仰角，方位角
  plt.show()
```

```python
if __name__ == '__main__':
  b = np.array([[1], [2]])
  A = np.array([[1, -2], [1, -2]])
  x0,x1,slope=show_gradient_descent(A, b)
  draw_gradient_descent(A, b,x0,x1,slope)
```
